```python
#Ejercicio 1
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def biseccion(a, b, tol=1e-2):
    if f(a) * f(b) >= 0:
        print("El intervalo [a, b] no contiene una raíz")
        return None
    c = a
    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c

# Intervalos dados
intervalos = [(0, 1), (1, 3.2), (3.2, 4)]

for intervalo in intervalos:
    a, b = intervalo
    raiz = biseccion(a, b)
    print(f"La raíz aproximada en el intervalo [{a}, {b}] es {raiz:.4f}")
```

```python
import math

# Datos dados
s0 = 300
m = 0.25
k = 0.1
g = 9.81

def s(t):
    return s0 - (m * g / k) * t + ((m**2 * g) / k**2) * (1 - math.exp(-k * t / m))

def ds_dt(t):
    return (-m * g / k) + ((m**2 * g) / k**2) * (math.exp(-k * t / m)) * (k / m)

def newton_raphson(func, deriv, x0, tol=0.01):
    x = x0
    while abs(func(x)) > tol:
        x = x - func(x) / deriv(x)
    return x

t0 = 1

tiempo_impacto = newton_raphson(s, ds_dt, t0)
print("El tiempo que tarda el objeto en golpear el piso es:", round(tiempo_impacto, 3), "segundos")
```
